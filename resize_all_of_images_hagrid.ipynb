```python
import shutil
import cv2
import os
from google.colab.patches import cv2_imshow
list_img = {
    
}
targetSize = 512

def processing(folder_path):
  cnt = 0
  for img_path in os.listdir(folder_path):
    image = cv2.imread(os.path.join(folder_path,img_path),3)
    #os.remove(os.path.join(folder_path,img_path))
    image = cv2.resize(image,(targetSize,targetSize))
    cv2.imwrite('/content/drive/MyDrive/hagrid_dataset/images/like'+img_path,image)
    cnt +=1
    print("NUMBER OF ITERATION={}, GESTURE_NAME={}".format(cnt,folder_path.split('/')[-1]))


path = '/content/drive/MyDrive/hagrid_dataset/dataset/like'

processing(path)
```
